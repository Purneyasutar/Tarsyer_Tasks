```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_path = "C:/Users/kaush/Downloads/Original Image.jpg"
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)


_, binary_image = cv2.threshold(original_image, 127, 255, cv2.THRESH_BINARY)

invert = cv2.bitwise_not(binary_image) 

kernel_size = 5
kernel = np.ones((kernel_size, kernel_size), np.uint8)


erosion_result = cv2.erode(invert, kernel, iterations=1)

dilation_result = cv2.dilate(invert, kernel, iterations=1)

opening_result = cv2.morphologyEx(invert, cv2.MORPH_OPEN, kernel)



plt.figure(figsize=(20, 5))


plt.subplot(1, 4, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(erosion_result, cmap='gray')
plt.title('Erosion Result')

plt.subplot(1, 4, 3)
plt.imshow(dilation_result, cmap='gray')
plt.title('Dilation Result')

plt.subplot(1, 4, 4)
plt.imshow(opening_result, cmap='gray')
plt.title('Opening Result')

plt.show()
```
